```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
# let implement group by function
# survival count by gender
df.groupby('sex')['survived'].count()
# this give how many women and men were on board
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# grouping by pclass
df.groupby('class')['survived'].count()
```

```python
# similary find out passenger embarked from town
df.groupby('embark_town')['survived'].count()
```

```python
# age analysis using .agg()
```

```python
df.groupby(['sex'])['age'].agg(['mean','min','max'])
```

```python
# find fare for every passenger class
df.groupby(['class'])['fare'].agg(['mean','min','max'])
```

```python
# how many male and female died and survivaed
df.groupby('sex')['survived'].value_counts()
```

```python
df.groupby('sex').agg(
    total_passengers=('survived','count'),
    avg_age = ('age','mean'),
    survival_rate = ('survived','mean')
)
```
